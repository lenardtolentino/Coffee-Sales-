```python
# revenue by month

x = data.groupby(['MonthName'])['Revenue'].sum().reset_index().sort_values(by = 'Revenue', ascending = True).round(1)

def thousands(x, pos):
    return '${:,.0f}k'.format(x)
    
formatter = FuncFormatter(thousands)

plt.figure(figsize = (9,5))
plt.plot(x['MonthName'], x['Revenue'], marker= 'o', color = '#00003f')

for i, txt in enumerate(x['Revenue']):
 plt.annotate(thousands(txt, None), (x['MonthName'].iloc[i], x['Revenue'].iloc[i]), textcoords = 'offset points', xytext = (0, 10), ha = 'center', fontsize = 10)

plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid()

plt.gca().yaxis.set_major_formatter(formatter)

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
x = data.groupby(['MonthName'])['transaction_id'].count().reset_index().sort_values(by = 'transaction_id', ascending = True)
```

```python
x = data.groupby(['WeekdayName'])['transaction_id'].count().reset_index().sort_values(by = 'transaction_id', ascending = True)
plt.figure(figsize = (10,5))
plt.bar(x['WeekdayName'], x['transaction_id'], color = '#00003f', width = 0.5)
plt.title("Transaction by day of Week")
plt.xlabel('day')
plt.ylabel("transaction")
plt.show()
```

```python
x = data.groupby(['Hours'])['transaction_id'].count().reset_index().sort_values(by = 'transaction_id', ascending = False)
plt.figure(figsize = (10,5))
plt.bar(x['Hours'], x['transaction_id'], color = '#00003f', width = 0.3)
plt.title("Transaction by Hour of Day")
plt.xlabel('Hours')
plt.ylabel("transaction")
plt.show()
```

```python
x=data.groupby(['product_category'])['transaction_id'].count().reset_index().sort_values(by = 'transaction_id', ascending = False)
plt.figure(figsize = (15,7))
bars = plt.barh(x['product_category'], x['transaction_id'], height = 0.5, color = '#00003f')

for bar in bars:
    plt.text(
        bar.get_width() + 5,
        bar.get_y() + bar.get_height() / 2,
        f'{bar.get_width()}',
        va = 'center',
        ha = 'left',
        fontsize = 10,
        color = '#00003f',
        
    )



plt.title('Transaction Per Product Category')
plt.show()
```

```python
x = data.groupby('product_type').agg(transaction_count=('transaction_id', 'count'), total_rev=('Revenue', 'sum')).reset_index()
top_15 = x.sort_values(by = 'transaction_count', ascending = False).head(15)
print(top_15.to_string(index = False))
```
